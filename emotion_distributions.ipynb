# Emotional Analysis

```python
import pandas as pd
import re
import matplotlib.pyplot as plt
data = pd.read_csv('emotions_data.csv')
data.head()
import numpy as np
# Apply regular expression to extract words from 'emotions' column
for i in range(len(data)):
    data.at[i, 'emotions'] = re.findall('\w+', data.at[i, 'emotions'])

data['emotion_scores'] = data['emotion_scores'].apply(lambda x: list(map(float, re.findall(r'\d+\.\d+', x))) if pd.notnull(x) else [])

```

# Average Emotion Score per Headline

```python
dictionary = {'Concern':0,'Skepticism':0,'Excitement':0,'Optimism':0}
for k in range(len(data)):
    for i in range(4):
        if data.iloc[k]['emotions'][i] == 'Concern':
            dictionary['Concern'] += data.iloc[k]['emotion_scores'][i]
        if data.iloc[k]['emotions'][i] == 'Skepticism':
            dictionary['Skepticism'] += data.iloc[k]['emotion_scores'][i]
        if data.iloc[k]['emotions'][i] == 'Excitement':
            dictionary['Excitement'] += data.iloc[k]['emotion_scores'][i]
        if data.iloc[k]['emotions'][i]== 'Optimism':
            dictionary['Optimism'] += data.iloc[k]['emotion_scores'][i]
values = list(map(lambda x: x / 10000, list(dictionary.values())))
keys = list(dictionary.keys())
sorted_data = sorted(zip(keys, values), key=lambda x: x[1], reverse=True)
sorted_keys, sorted_values = zip(*sorted_data)
bar_colors = plt.cm.viridis(np.linspace(0, 1, len(sorted_keys)))

plt.figure()
plt.bar(sorted_keys, sorted_values, color=bar_colors)
plt.ylabel('Average Emotion Score')
plt.xlabel('Emotion')
plt.title('Average Emotion Score Per Headline')
plt.show()
```

```python
conDictionary = {5:0,6:0,7:0,8:0,9:0,10:0,11:0}
skeptDictionary = {5:0,6:0,7:0,8:0,9:0,10:0,11:0}
optDictionary = {5:0,6:0,7:0,8:0,9:0,10:0,11:0}
ExDictionary = {5:0,6:0,7:0,8:0,9:0,10:0,11:0}
mayCount = 0
juneCount = 0
julyCount = 0
augCount = 0
sepCount = 0
octCount = 0
novCount = 0
for k in range(len(data)):
    for i in range(4):
        if data.iloc[k]['month'] == 5:        
            if data.iloc[k]['emotions'][i] == 'Skepticism':
                skeptDictionary[5] += data.iloc[k]['emotion_scores'][i]
            if data.iloc[k]['emotions'][i] == 'Excitement':
                ExDictionary[5] += data.iloc[k]['emotion_scores'][i]
            if data.iloc[k]['emotions'][i] == 'Optimism':
                optDictionary[5] += data.iloc[k]['emotion_scores'][i]
            if data.iloc[k]['emotions'][i] == 'Concern':
                conDictionary[5] += data.iloc[k]['emotion_scores'][i]
            mayCount += 1
        if data.iloc[k]['month'] == 6:        
            if data.iloc[k]['emotions'][i] == 'Skepticism':
                skeptDictionary[6] += data.iloc[k]['emotion_scores'][i]
            if data.iloc[k]['emotions'][i] == 'Excitement':
                ExDictionary[6] += data.iloc[k]['emotion_scores'][i]
            if data.iloc[k]['emotions'][i] == 'Optimism':
                optDictionary[6] += data.iloc[k]['emotion_scores'][i]
            if data.iloc[k]['emotions'][i] == 'Concern':
                conDictionary[6] += data.iloc[k]['emotion_scores'][i]
            juneCount += 1
        if data.iloc[k]['month'] == 7:        
            if data.iloc[k]['emotions'][i] == 'Skepticism':
                skeptDictionary[7] += data.iloc[k]['emotion_scores'][i]
            if data.iloc[k]['emotions'][i] == 'Excitement':
                ExDictionary[7] += data.iloc[k]['emotion_scores'][i]
            if data.iloc[k]['emotions'][i] == 'Optimism':
                optDictionary[7] += data.iloc[k]['emotion_scores'][i]
            if data.iloc[k]['emotions'][i] == 'Concern':
                conDictionary[7] += data.iloc[k]['emotion_scores'][i]
            julyCount += 1
        if data.iloc[k]['month'] == 8:        
            if data.iloc[k]['emotions'][i] == 'Skepticism':
                skeptDictionary[8] += data.iloc[k]['emotion_scores'][i]
            if data.iloc[k]['emotions'][i] == 'Excitement':
                ExDictionary[8] += data.iloc[k]['emotion_scores'][i]
            if data.iloc[k]['emotions'][i] == 'Optimism':
                optDictionary[8] += data.iloc[k]['emotion_scores'][i]
            if data.iloc[k]['emotions'][i] == 'Concern':
                conDictionary[8] += data.iloc[k]['emotion_scores'][i]
            augCount += 1
        if data.iloc[k]['month'] == 9:        
            if data.iloc[k]['emotions'][i] == 'Skepticism':
                skeptDictionary[9] += data.iloc[k]['emotion_scores'][i]
            if data.iloc[k]['emotions'][i] == 'Excitement':
                ExDictionary[9] += data.iloc[k]['emotion_scores'][i]
            if data.iloc[k]['emotions'][i] == 'Optimism':
                optDictionary[9] += data.iloc[k]['emotion_scores'][i]
            if data.iloc[k]['emotions'][i] == 'Concern':
                conDictionary[9] += data.iloc[k]['emotion_scores'][i]
            sepCount += 1
        if data.iloc[k]['month'] == 10:        
            if data.iloc[k]['emotions'][i] == 'Skepticism':
                skeptDictionary[10] += data.iloc[k]['emotion_scores'][i]
            if data.iloc[k]['emotions'][i] == 'Excitement':
                ExDictionary[10] += data.iloc[k]['emotion_scores'][i]
            if data.iloc[k]['emotions'][i] == 'Optimism':
                optDictionary[10] += data.iloc[k]['emotion_scores'][i]
            if data.iloc[k]['emotions'][i] == 'Concern':
                conDictionary[10] += data.iloc[k]['emotion_scores'][i]
            octCount += 1
        if data.iloc[k]['month'] == 11:        
            if data.iloc[k]['emotions'][i] == 'Skepticism':
                skeptDictionary[11] += data.iloc[k]['emotion_scores'][i]
            if data.iloc[k]['emotions'][i] == 'Excitement':
                ExDictionary[11] += data.iloc[k]['emotion_scores'][i]
            if data.iloc[k]['emotions'][i] == 'Optimism':
                optDictionary[11] += data.iloc[k]['emotion_scores'][i]
            if data.iloc[k]['emotions'][i] == 'Concern':
                conDictionary[11] += data.iloc[k]['emotion_scores'][i]
            novCount += 1
```

```python
list_of_dicts = [conDictionary, ExDictionary, skeptDictionary, optDictionary]

maySum = sum([i[5] for i in list_of_dicts])
juneSum = sum([i[6] for i in list_of_dicts])
julySum = sum([i[7] for i in list_of_dicts])
augSum = sum([i[8] for i in list_of_dicts])
sepSum = sum([i[9] for i in list_of_dicts])
octSum = sum([i[10] for i in list_of_dicts])
novSum = sum([i[11] for i in list_of_dicts])

sums = [maySum, juneSum, julySum, augSum, sepSum, octSum, novSum]

keys = sorted(set(key for d in list_of_dicts for key in d.keys()))
values = [[d.get(key, 0) for key in keys] for d in list_of_dicts]

values_percentage = np.array(values) / np.sum(values, axis=0) * 100

colors = np.random.rand(len(list_of_dicts), 3)

fig, ax = plt.subplots()
bottom = np.zeros(len(keys))
names = ['Concern', 'Excitement', 'Skepticism', 'Optimism']

for i, dataset_values in enumerate(values_percentage):
    ax.bar(keys, dataset_values, bottom=bottom, label=names[i], color=colors[i])
    bottom += dataset_values

plt.xlabel('Month')
plt.ylabel('Values (%)')
plt.title('Monthly Distribution of Emotion Scores')

plt.xticks(keys, ['May', 'June', 'July', 'August', 'September', 'October', 'November'],rotation = 45)

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```
