```python
import pandas as pd
from transformers import pipeline

classifier = pipeline("zero-shot-classification",
                      model="facebook/bart-large-mnli")
```

```python
data = pd.read_csv('clean_titles_10k.csv')
```

```python
data
```

```python
test_data = data[0:500]
```

```python
labels = ["Excitement", "Optimism",'Concern', 'Skepticism']
a_review = 'Has ChatGPT failed, does it signal failure of Artificial Intelligence as'
res = classifier(a_review, labels, multi_label=True)
print(res)
```

```python
from tqdm import tqdm
```

```python
results = []
emotion_scores = []

for idx, item in tqdm(test_data.iterrows()):
    res = classifier(item['cleaned_title'], labels, multi_label = True)

    results.append( res['labels'])
    emotion_scores.append( res['scores'])
```

```python
test_data['Predicted Emotion'] = results
```
