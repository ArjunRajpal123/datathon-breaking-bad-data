```python
import pandas as pd
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
```

```python
data = pd.read_csv('BreakingBadData/Dataset_3.5k.csv')
```

```python
test_data = data[0:199]
```

```python
test_data
```

```python
map = {-1: 'negative', 0: 'neutral', 1: 'positive'}
test_data.loc[:, 'Manual Sentiment'] = test_data['Manual Sentiment'].replace(map)
test_data
```

```python
labels = ["positive", "negative", "neutral"]
a_review = "Fake access to malicious AI tool 'WormGPT' for sale on dark web - Edge Middle East"
res = classifier(a_review, labels)
print(res)
```

```python
from sklearn.metrics import accuracy_score
from tqdm import tqdm
```

```python
results = []
targets = []

for idx, item in tqdm(test_data.iterrows()):
    res = classifier(item['translated_title'], labels)

    results.append( res['labels'][0])
    targets.append( item['Manual Sentiment'])

accuracy = accuracy_score(results, targets)
```

```python
accuracy
```

```python
test_data['Predicted Sentiment'] = results
```
